```python
#將每張圖檔與對應的分數塞入字典，作排序(字母從小到大)
dic={}
with open('classify_list', 'r') as f:
    for list in f.readlines():
        filename = list.strip().split('|')[1]
        filescore = list.strip().split('|')[0]
        #print filename + '|' + filescore
        if not filename in dic:
            dic[filename] = filescore
    dict= sorted(dic.iteritems(), key=lambda d:d[0])
```

```python
#將字典內的key與value取出，寫入新檔案
for i in range(0,1441):
    f = open('processed_list', 'a')
    f.write(dict[i][0] + '|' + dict[i][1] + '\n')
    f.close() 
    print imgList
```
